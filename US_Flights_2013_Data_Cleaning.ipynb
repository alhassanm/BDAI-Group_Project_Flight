```python
# first we import our libraries
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib as plt
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('flight_2013.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
for i in range(len(df)):
  number = df['DepDelay'][i]
  if np.isnan(df['DepDel15'][i]):
    if number < 15:
      df['DepDel15'][i] = 0
    elif number >= 15:
      df['DepDel15'][i] = 1
```

```python
df.dropna(inplace = True)
```

```python
df['DayTime_Dep'] = ''
df['DayTime_Arr'] = ''
for i in range(len(df)):
    if( len(str(int(df['CRSDepTime'][i]))) <= 2):
      df['DayTime_Dep'][i] = datetime.datetime(df['Year'][i], df['Month'][i], df['DayofMonth'][i], 00, int(str(int(df['CRSDepTime'][i]))[-2:]) )
    else:
      df['DayTime_Dep'][i] = datetime.datetime(df['Year'][i], df['Month'][i], df['DayofMonth'][i], int(str(int(df['CRSDepTime'][i]))[:-2]), int(str(int(df['CRSDepTime'][i]))[-2:]) ) 

    if( len(str(int(df['CRSArrTime'][i]))) <= 2):
      df['DayTime_Arr'][i] = datetime.datetime(df['Year'][i], df['Month'][i], df['DayofMonth'][i], 00, int(str(int(df['CRSArrTime'][i]))[-2:]) ) 
    else:
      df['DayTime_Arr'][i] = datetime.datetime(df['Year'][i], df['Month'][i], df['DayofMonth'][i], int(str(int(df['CRSArrTime'][i]))[:-2]), int(str(int(df['CRSArrTime'][i]))[-2:]) ) 
```
